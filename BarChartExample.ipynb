```python
""" The frequency bar chart of a six sided die: This chart is built by using mathplotlib"""
# Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

def add_labels(x,y):#accept 2 list arguments 
    for i in range(len(x)):
        plt.text(i+1,y[i],y[i],ha = 'center')# First two arguments are coordinates of the plot, the third one is the value and the last one
                                            # is for horizontal alignment.
    
rolls=[random.randrange(1,7) for item in range(int(input("how many times do you wanna roll the die?")))] # List comprehension to create a list of 6000 numbers between the range 1-6

values, frequencies=np.unique(rolls,return_counts=True) # creating a narray of unique values and their counts in the list

tittle= "Rolling a six-sided die {:,} times".format(len(rolls))
c = ['red', 'yellow', 'black', 'blue', 'orange','green'] # for the color of bars
plt.ylabel("Frequencies")
plt.xlabel("Die Value")
plt.title(tittle)
plt.bar(values,frequencies,color=c) # Create a bar colored-chart
add_labels(values,frequencies) # Invoke the function for bar values
plt.ylim(0, max(frequencies)*1.1) # To set the y-limits of the current axes. *By default is the nearest to number. Here %10 more of top number
plt.show()
```
